```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10,10,0.01)
def f(x):
    return 0.5*x**4 - 3*x**3
plt.ylim(-100,100)
plt.xlim(-5,7)
plt.plot(x,f(x))
```

```python
x_old = 0
x_new = 4
eps = 0.001
precision = 0.00001

def df(x):
    return 2 * x**3 - 9 * x**2

while abs(x_new - x_old) > precision:
    x_old = x_new
    x_new = x_old - eps * df(x_old)
    
print(x_new)
```

```python
from sklearn import datasets
from sklearn.linear_model import LinearRegression
```

```python
diabetes=datasets.load_diabetes()
```

```python
diabetes.keys()
```

```python
print(diabetes.DESCR)
```

```python
X_age = diabetes.data[:,0] # 나이
X_bmi = diabetes.data[:,2] # BMI
X_p = diabetes.data[:,3] # 혈압
Y = diabetes.target
```

```python
print(diabetes.data.shape, diabetes.target.shape)
print(X_age.shape)
print(X_bmi.shape)
print(X_p.shape)
```

```python
X_1 = np.array(X_age)
Y_1 = np.array(Y)
plt.plot(X_1, Y_1,'o')
```

```python
a, b = 7, 6
yh = a*X_1 + b
np.mean((yh-Y_1)**2)
```

```python
def f(a,b,X_i):
    return a * X_i + b

def MSE1(a,b,Y_1):
    yh = a * X_1 + b
    return np.mean((yh-Y_1)**2)
```

```python
eps = 0.1
n = 0;
while(n<100000):
    if(MSE1(a+eps,b,Y_1)>MSE1(a-eps,b,Y_1)):
        a = a - eps
    else:
        a = a + eps
    
    if(MSE1(a,b+eps,Y_1)>MSE1(a,b-eps,Y_1)):
        b = b - eps
    else:
        b = b + eps
    n = n + 1
```

```python
a,b
```

```python
plt.plot(X_1,Y_1,'o')
plt.plot(X_1,f(a,b,X_1))
```

```python
X__1 = np.array([X_age])
Y__1 = np.array([Y])

X___1 = np.array([[i] for i in X_age])
Y___1 = np.array([[i] for i in Y])
plt.plot(X__1,Y__1,'o')
```

```python
f_1 = LinearRegression()
f_1.fit(X___1,Y___1)
```

```python
f_1.coef_, f_1.intercept_
```

```python
plt.plot(X__1,Y__1,'o')
plt.plot(X___1,f_1.predict(X___1))
```
